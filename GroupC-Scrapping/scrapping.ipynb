```python
title_list =[]
rating_list = []
price_list = []

for pg_num in range(1,51):
  req = requests.get(f'https://books.toscrape.com/catalogue/page-{pg_num}.html')
  soup_pg = bs4.BeautifulSoup(req.text)

  for item in soup_pg.select('article h3 a'):
    title_list.append(item['title'])

  for n in range(0,60,3):
     rating_list.append(soup_pg.select('article p')[n]['class'][1])

  for n in range(1,60,3):
     price_list.append(soup_pg.select('article p')[n].string.replace('Â£',''))
```

```python
dic = dict(zip(title_list,zip(rating_list, price_list)))
df = pd.DataFrame.from_dict(dic, orient='index')
df
```

```python
new_df = pd.DataFrame()  # empty dataframe
index_num = 0  # initialize index for new_df

#here we can take the range from 1 to 51 but it takes too much time therefore we take only 1 to 5 for now
for pg_num in range(1, 5):
    req = requests.get(f'https://books.toscrape.com/catalogue/page-{pg_num}.html')
    soup_pg = bs4.BeautifulSoup(req.text)
    book_info_list = soup_pg.select('article h3 a')  # list of all books on that page

    for n in range(0, 20):
        res = requests.get('http://books.toscrape.com/catalogue/' + book_info_list[n]['href'])  # book's individual page
        soup = bs4.BeautifulSoup(res.text, 'lxml')

        book_features_column_list = soup.select('tr th')  # all feature headings like price, rating
        book_column_values = soup.select('tr td')  # corresponding values for those features

        column_names = [item.text for item in book_features_column_list]  # list of feature names (e.g., 'Price', 'Availability')
        column_values = [item.text for item in book_column_values]  # list of feature values

        d = dict(zip(column_names, column_values))  # creating a dictionary with feature names as keys and values as values
        df = pd.DataFrame(d, index=[index_num + n])  # creating a DataFrame for the current book

        # Append the DataFrame to the new DataFrame using pd.concat
        new_df = pd.concat([new_df, df], ignore_index=True)  # Use pd.concat instead of append

    index_num += 20  # increment the index for the next page
```
