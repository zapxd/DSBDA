```python
subset1 = df[['age', 'workclass', 'education', 'income']]

print(subset1)
```

```python
# Subset 2: gender, race, Category, capital-gain
subset2 = df[['age', 'occupation', 'gender', 'capital-gain']]

print(subset2)
```

```python
# Merge Subset1 and Subset2 based on 'age'
merged_data = pd.merge(subset1, subset2, on='age')
print("\nMerged Subset1 and Subset2 (on 'age'):")
print(merged_data)
```

```python
# Merge Subset1 and Subset2 based on different columns
merged_different_cols = pd.merge(subset1, subset2, left_on='workclass', right_on='occupation')

print("\nMerged Subset1 and Subset2 (left_on='workclass', right_on='occupation'):")
print(merged_different_cols)
```

```python
# Sorting merged_data in descending order wrt 'capital-gain'
merged_data.sort_values(by=['capital-gain'],ascending=False)
```
