```python
import pandas as pd
```

```python
df = pd.read_csv('dataset_Facebook.csv',delimiter=';')
```

```python
df.describe()
```

```python
# First subset: Like and Share and Category
df_subset_1 = df[['Category', 'like', 'share']]
df_subset_1
```

```python
# second subset: Comment and Type
df_subset_2 = df[['Category', 'comment', 'Type']]
df_subset_2
```

```python
merged_data  = pd.merge(df_subset_2, df_subset_1, left_on='comment', right_on= 'like')
merged_data
```

```python
merged_on_category = pd.merge(df_subset_1, df_subset_2, on='Category')
merged_on_category
```

```python
# Sorting merged_data in descending order wrt 'Like'
merged_data.sort_values(by=['like'],ascending=False)
```
